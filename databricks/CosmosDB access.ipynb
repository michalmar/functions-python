```
from faker import Faker
fake = Faker()
first_names = [fake.first_name() for i in range(1000)]
last_names = [fake.last_name() for i in range(1000)]
emails = [fake.ascii_company_email() for i in range(1000)]


```

```

import random
import string
import sys
import faker



def random_string(length):
    return ''.join(random.choice(string.ascii_letters) for _ in range(length))
# functions generates random guid
def random_guid():
    return random_string(8)+"-"+random_string(4)+"-"+random_string(4)+"-"+random_string(4)+"-"+random_string(12)
def random_int(min, max):
    return random.randint(min, max)
def random_float(i):
    return random.uniform(i-1, i)
# def random_float(min, max):
#     return random.uniform(min, max)
def random_date():
    return '2021-{:02d}-{:02d}'.format(random_int(1, 12), random_int(1, 28))

def random_first_name(l):
#     fake = Faker()
#     return fake.first_name()
    return random.choice(l)
def random_last_name(l):
#     fake = Faker()
#     return fake.last_name()
    return random.choice(l)
def random_email(l):
#     fake = Faker()
#     return fake.ascii_company_email()
    return random.choice(l)
def random_phone():
    return random_string(10)
def random_city():
    return random_string(10)
def random_state():
    return random_string(10)
def random_country():
    return random_string(10)



def random_customer(n):
    customer_attr_list= [random_guid(), random_first_name(first_names), random_last_name(last_names), random_email(emails), random_date(), random_date()]
    for i in range(n):
        customer_attr_list.append(random_float(1.0))
    return tuple(customer_attr_list)
    
    
```

```
from pyspark.sql.types import StructType,StructField, StringType, IntegerType, DateType, FloatType
# number of customer's decimal parameters
par_n = 200

# number of customers
cust_n = 1000*1000 #M

# generate customer record headers
struct = [StructField("id",StringType(),True),
        StructField("firstname",StringType(),True), 
        StructField("lastname",StringType(),True), 
        StructField("email",StringType(),True),
        StructField("date1", StringType(), True), 
        StructField("date2", StringType(), True)]

for i in range(par_n):
    struct.append(StructField(f"feat{i}",FloatType(),True))
    
schema = StructType(struct)

# generate customer records
data2 = []
for i in range(cust_n):
    data2.append(random_customer(par_n))
    
# create datframe
df = spark.createDataFrame(data=data2,schema=schema)
# df.printSchema()
# df.show(truncate=False)

display(df.take(4))
print(f"Generated {cust_n} customers (with {par_n} additional parameters), stored in Spark dataframe.")
```
